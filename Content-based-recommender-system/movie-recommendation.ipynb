# Content Based Recommendation Systems

A content-based recommendation system is a type of recommendation system that suggests items to users based on the characteristics and features of those items, as well as the user's profile or preferences. This approach is particularly useful when you have detailed information about the content of items and the preferences of users.

```python
number_rating=ratings_with_books.groupby('title')['ratings'].count().reset_index()
```

```python
number_rating.rename(columns={'ratings':'number_of_ratings'},inplace=True)
```

```python
final_rating=ratings_with_books.merge(number_rating,on='title')

final_rating.shape
```

```python
final_rating=final_rating[final_rating['number_of_ratings']>=10]
```

```python
final_rating.head(3)
```

```python
final_rating.drop_duplicates(['user-id','title'],inplace=True)
```

```python
final_rating['title'].value_counts()
```

```python
def remove_space(x):
    x= str(x)
    return x.replace(" ", '')
```

```python
final_rating['author'] = final_rating['author'].apply(remove_space)
final_rating['publisher'] = final_rating['publisher'].apply(remove_space)

```

```python
final_rating
```

```python
final_rating.reset_index(inplace=True)
```

```python
final_rating['ratings'] = final_rating['ratings'].astype(str)
final_rating['number_of_ratings'] = final_rating['number_of_ratings'].astype(str)
```

```python
final_rating["combined"] = final_rating['ratings'] + ' ' + final_rating['author']  + ' ' + final_rating['publisher'] + ' ' + final_rating['number_of_ratings']
final_rating.drop(['ratings','author','publisher','number_of_ratings'],axis=1,inplace=True)
final_rating.head()
```

```python

final_rating
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel



# TF-IDF Vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(final_rating["combined"])

# Compute cosine similarities
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
movie_title = final_rating['title']
indices = pd.Series(final_rating.index, index=final_rating['title'])

# print("Search recommendation for")
# inpput_movie= input()
# idx = indices.get(inpput_movie)  
# sim_scores = list(enumerate(cosine_similarities[idx]))
# sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
# sim_scores = sim_scores[0:10]
# # print(sim_scores)
# movie_indices = [i[0] for i in sim_scores]
# print(movie_indices,'movie_indices')
# movie_title.iloc[movie_indices]
```

```python
def content_recommender(title):
    idx = indices.get(title)
    if idx is None:
        print("Title not found in indices dictionary.")
        return None
    sim_scores = list(enumerate(cosine_similarities[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1].max(), reverse=True)
    sim_scores = sim_scores[1:11]  # Exclude the input book itself
    movie_indices = [i[0] for i in sim_scores]
    return movie_title.iloc[movie_indices]

# Example usage
recommended_books = content_recommender("Midnight Whispers (Cutler)")
print(recommended_books)
```
